```python
# 1. Create a class BankAccount with the following: a) Attributes: account_holder, account_number, and balance. b) Methods: deposit(amount), withdraw(amount), and get_balance(). c) withdraw() should print a message and do nothing if the amount exceeds the current balance. d) A __str__ method that returns a readable summary of the account. Create two BankAccount objects and perform a few deposits and withdrawals on each.
class BankAccount:
    def __init__(self, name, acc_no, balance):
        self.name = name
        self.acc_no = acc_no
        self.balance = balance

    def deposit(self, amount):
        self.balance=self.balance+amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance= self.balance-amount

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"{self.name} {self.acc_no} Balance: {self.balance}"
```

```python
a1.deposit(1000)
a1.withdraw(2000)
a2.deposit(500)
a2.withdraw(2000)
```

```python
# 2. Create a class Student with attributes name, roll_no, and marks (a list of 5 subject marks). Add methods: a) calculate_percentage() — returns the average of all marks. b) get_grade() — returns 'A' for >= 80, 'B' for >= 60, 'C' for >= 40, and 'Fail' otherwise. c) display() — prints all details including percentage and grade. Create a list of 3 Student objects and call display() on each.
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        total = 0
        for m in self.marks:
            total = total + m
        return total / 5

    def get_grade(self):
        per = self.calculate_percentage()

        if per >= 80:
            return "A"
        elif per >= 60:
            return "B"
        elif per >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Percentage:", self.calculate_percentage())
        print("Grade:", self.get_grade())
```

```python
s1 = Student("Ravi", 101, [80, 85, 90, 75, 70])
s2 = Student("Priya", 102, [60, 65, 70, 55, 50])
s3 = Student("Amit", 103, [30, 35, 40, 45, 50])
students = [s1, s2, s3]
for s in students:
    s.display()
```

```python
# 4. Build a class SecureVault that stores a private __pin and a private __data dictionary. It should have: a) A method add_entry(pin, key, value) that adds to __data only if the pin matches. b) A method get_entry(pin, key) that returns the value only if the pin matches. c) Any wrong pin attempt should print 'Access Denied' and return nothing.
class SecureVault:
    def __init__(self):
        self.__pin = 1234
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data[key]
        else:
            print("Access Denied")
```
