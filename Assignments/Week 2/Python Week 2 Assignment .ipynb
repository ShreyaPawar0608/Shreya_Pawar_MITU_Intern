```python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    result = {
        "word_count": len(words),
        "char_count": len(sentence.replace(" ", "")),
        "unique_words": len(set(words)),
        "most_frequent_word": max(words, key=words.count)
    }

    return result

s = "The quick brown fox jumps over the lazy dog the fox"

print(analyze_sentence(s))
```

```python
text = "A man a plan a canal Panama"

# a) Reverse string
rev = ""
for i in text:
    rev = i + rev
print("Reverse =", rev)

# b) Palindrome check
s = text.replace(" ", "").lower()
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

# c) Replace vowels with *
new = ""
for i in text:
    if i.lower() in "aeiou":
        new += "*"
    else:
        new += i
print("New String =", new)

# d) Count vowels
for v in "aeiou":
    print(v, "=", text.lower().count(v))
```

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
def list_ops(nums, nested):

    # a) Remove duplicates
    unique = []
    for i in nums:
        if i not in unique:
            unique.append(i)
    print("Without Duplicates:", unique)

    # b) Elements at even indices squared
    squared = [nums[i]**2 for i in range(0, len(nums), 2)]
    print("Squared Even Indices:", squared)

    # c) Flatten nested list
    flat = [x for sublist in nested for x in sublist]
    print("Flattened List:", flat)


nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

list_ops(nums, nested)
```

```python
employee = ("E042", "Riya Sharma", "Data Science", 72000, "Pune")

# a) Unpack tuple
emp_id, name, department, salary, city = employee

print(emp_id)
print(name)
print(department)
print(salary)
print(city)

# b) * unpacking
first, *middle, last = employee

print("First =", first)
print("Middle =", middle)
print("Last =", last)

# c) Update salary
def update_salary(emp_tuple, new_salary):
    return (emp_tuple[0], emp_tuple[1], emp_tuple[2],
            new_salary, emp_tuple[4])

new_emp = update_salary(employee, 80000)

print("Updated Tuple =", new_emp)
```

```python
batch_a = {"Aarav", "Priya", "Karan", "Meera", "Sahil"}
batch_b = {"Karan", "Meera", "Tanvi", "Rohan"}
batch_c = {"Priya", "Aarav"}

# a) Common students
print("Both batches:", batch_a & batch_b)

# b) In batch_a but not in batch_b
print("Only in batch_a:", batch_a - batch_b)

# c) All unique students
print("All students:", batch_a | batch_b)

# d) Check subset
print("Batch_c is subset of batch_a:", batch_c.issubset(batch_a))
```

```python
def inventory_manager(records):
    inventory = {}

    for product, category, quantity, price in records:

        if category not in inventory:
            inventory[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }

        inventory[category]["total_items"] += quantity
        inventory[category]["total_value"] += quantity * price
        inventory[category]["products"].append(product)

    return inventory


records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20)
]

print(inventory_manager(records))
```

```python
# a) power_all function
def power_all(*args, exp=2):
    return [x ** exp for x in args]

print(power_all(2, 3, 4))
print(power_all(2, 3, 4, exp=3))

# b) profile function
def profile(**kwargs):
    for k, v in kwargs.items():
        print(k, ":", v)

profile(name="Arjun", age=20, city="Pune")

# c) Sort by CGPA in descending order
students = [
    {"name": "Arjun", "cgpa": 8.4},
    {"name": "Sanya", "cgpa": 9.1},
    {"name": "Dev", "cgpa": 7.8}
]

sorted_students = sorted(students,
                         key=lambda x: x["cgpa"],
                         reverse=True)

print(sorted_students)
```

```python
import time
import math

# a) Closure
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print(triple(4))
print(quintuple(4))


# b) Timer decorator
def timer(func):
    def wrapper(n):
        start = time.time()
        result = func(n)
        end = time.time()
        print("Time:", end - start)
        return result
    return wrapper

@timer
def slow_sum(n):
    s = 0
    for i in range(1, n + 1):
        s += i * i
    return s

print(slow_sum(10000))


# c) Validate positive decorator
def validate_positive(func):
    def wrapper(x):
        if x <= 0:
            raise ValueError("Number must be positive")
        return func(x)
    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print(safe_sqrt(25))
```
