```python
# 1. Take two numbers as input from the user and print their sum, difference, product, and quotient — each on a separate line. Handle division by zero with an appropriate message.
num1=int(input("Enter number 1:"))
num2=int(input("Enter number 2:"))
sum=num1+num2
difference=num1-num2
product=num1*num2
quotient=num1//num2
print("Sum:",sum)
print("Difference:",difference)
print("Product:",product)
if num2!=0:
    print("Quotient:",quotient)
else:
    print("Division by zero is not allowed!")
```

```python
# 2. Accept a temperature in Celsius from the user, convert it to Fahrenheit and Kelvin, and print all three values with labels.
celsius=int(input("Enter temperature in celsius:"))
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15
print("\n--- Temperature Conversion ---")
print("Celsius:"    ,celsius)
print("Fahrenheit:" ,fahrenheit)
print("Kelvin:"     ,kelvin)
```

```python
# 3. Take a string as input and print: its uppercase version, its length, and the string reversed.
string=input("Enter any string:")
str=string.upper()
print("Uppercase Version:",str)
str=len(string)
print("String Length:",str)
str=string[::-1]
print("Reversed string:",str)
```

```python
# 4. Take three numbers as input and print the largest among them. Do not use the built-in max() function.
num1=int(input("Enter number 1:"))
num2=int(input("Enter number 2:"))
num3=int(input("Enter number 3:"))
if num1>num2 and num1>num3:
    print("Largest number is:",num1)
elif num2>num3:
    print("Largest number is:",num2)
else:
    print("Largest number is:",num3)
```

```python
# 5. Take a year as input and print whether it is a leap year or not.
year = int(input("Enter any year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
# 6. Build a simple calculator — take two numbers and an operator (+, -, *, /) as input, perform the operation, and print the result. Print an error message for invalid operators or division by zero.
num1=int(input("Enter number 1:"))
num2=int(input("Enter number 2:"))
operator=(input("Enter any operator(+,-,*,/):"))
if operator=='+':
    print("Addition:",num1+num2)
elif operator=='-':
    print("Subtraction:",num1-num2)
elif operator=='*':
    print("Multiplication:",num1*num2)
elif operator=='/':
    if num2!=0:
        print("Division:",num1/num2)
    else:
        print("Divison by zero is not allowed!")
else:
    print("Error: Invalid Operator!")
```

```python
# 7. Take a number as input and print its multiplication table from 1 to 10. Format each line as: 7 x 3 = 21
num=int(input("Enter any number between 1 to 10:"))
for i in range(1,11):
    print(num, 'x' ,i, '=' ,num*i)
```

```python
#8. Take a positive integer n as input and print all prime numbers from 2 to n
num=int(input("Enter any positive integer:"))
for i in range(2, num + 1):
    prime = True
    for j in range(2, i):
        if i % j == 0:
            prime = False
            break
    if prime:
        print(i)
```

```python
# 9. Keep accepting numbers from the user until they enter 0. Then print the total count of numbers entered, their sum, and their average. Do not include 0 in the calculations.
count = 0
sum = 0

while True:
    num = int(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    count = count + 1
    sum = sum + num

if count > 0:
    avg = sum / count
else:
    avg = 0

print("Count =", count)
print("Sum =", sum)
print("Average =", avg)
```

```python
# 10. Take n as input and print a triangle pattern of n rows made of asterisks, where row i has (2i - 1) stars centered appropriately.
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i), end="")
    print("*" * (2 * i - 1))
```
