## Python List Questions

```python
# Question 1: List Manipulation Basics
my_list=[10,20,30,40,50,60]
my_list[2]=35
my_list.append(70)
my_list.insert(0,5)
my_list.remove(40)
print(my_list)
```

```python
# Question 2: Sum of Even Numbers
l=[12, 7, 23, 48, 15, 66, 9]
sum=0
for i in l:
    if i%2==0:
        sum=sum+i
print("Sum of all even numbers from the list is",sum)
```

```python
# Question 3: Manual List Reversal
data = ['apple', 'banana', 'cherry', 'date', 'elderberry']
reversed_list = []
for i in range(len(data)  -1, -1, -1):
    reversed_list.append(data[i])
print(reversed_list)
```

```python
# Question 4: Finding the Maximum Element
l=[8, 2, 10, 5, 17, 3, 14]
largest=l[0]
for i in l:
    if i>largest:
        largest=i
print("Largest number in list values is",largest)
    
```

```python
# Question 5: Count Occurrences
items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red']
count=0
for color in items:
    if color=='red':
        count=count+1
print("Count of red in the list is",count)
```

```python
# Question 6: Remove All Occurrences of an Element
mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5]
for i in mixed_numbers:
    if i==5:
        mixed_numbers.remove(i)
print(mixed_numbers)
```

```python
# Question 7: List Palindrome Check
list1 = [1, 2, 3, 2, 1]
list2 = ['a', 'b', 'c', 'b', 'd']
for lst in [list1, list2]:
    rev = []
    for i in range(len(lst)-1, -1, -1):
        rev.append(lst[i])
    if lst == rev:
        print("Palindrome")
    else:
        print("Not a Palindrome")
```

```python
# Question 8: Merge and Sort Two Lists (Manual Sort)
list_a = [3, 1, 4]
list_b = [2, 5, 0]
combined = list_a + list_b
for i in range(len(combined)):
    for j in range(len(combined) - 1):
        if combined[j] > combined[j + 1]:
            temp = combined[j]
            combined[j] = combined[j + 1]
            combined[j + 1] = temp
print("Sorted List:", combined)
```

```python
# Question 9: Flaen a Nested List
nested_list = [1, [2, 3], 4, [5, 6, 7], 8]

flat_list = []

for i in nested_list:
    if type(i) == list:
        for j in i:
            flat_list.append(j)
    else:
        flat_list.append(i)

print(flat_list)
```

```python
# Question 10: Find Common Elements
list_p = [1, 2, 3, 4, 5]
list_q = [4, 5, 6, 7, 8]
common = []
for i in list_p:
    for j in list_q:
        if i==j:
            common.append(i)
print("Common elements:",common)    
```
