```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('banknotes.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# check for pattern through viz
```

```python
sns.pairplot(df, hue = 'Class');
```

```python
# check the distribution of target variable (countplot)
```

```python
x = df.drop('Class', axis = 1)
y = df['Class']
```

```python
sns.countplot(x = y);
```

```python
df['Class'].value_counts()
```

```python
# cross val
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    random_state=0, test_size=0.25)
```

```python
x_train.shape, y_train.shape
```

```python
# build the decision tree
# and also evaluate
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
# evaluate
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
x.columns
```

```python
dt.feature_importances_
```

```python
# plot the graph to viz feature importances
```

```python
plt.title("feature importances")
plt.xlabel('features')
plt.ylabel("importance")
plt.bar(x.columns, dt.feature_importances_);
```

```python
# plot the tree
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(16, 12))
plot_tree(dt, rounded= True, fontsize= 7, filled= True,
          class_names= ['Original', 'Fake'],
          feature_names= x.columns);
```
