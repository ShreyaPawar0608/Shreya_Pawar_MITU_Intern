```python
s = input("Enter a string: ")

c1 = input("Enter the first character to check: ")
c2 = input("Enter the second character to check: ")

f1 = False
f2 = False

for ch in s:
    if ch == c1:
        f1 = True
    if ch == c2:
        f2 = True

if f1:
    print(c1, "is present in",s)
else:
    print(c1, "is not present in",s)

if f2:
    print(c2, "is present in",s)
else:
    print(c2, "is not present in",s)
            
```
