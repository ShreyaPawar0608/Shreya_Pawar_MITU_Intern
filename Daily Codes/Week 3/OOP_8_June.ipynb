```python
class Person:
    
    def __init__(self,name,age,ht):
        self.name = name
        self.__age = age #private variable
        self.height = ht
        
    def show_info(self):
        return f"Name -> {self.name}, age -> {self.__age}, height -> {self.height}"
    
```

```python
p1=Person("Jatin",20,5.8)
p2=Person("Sachin",22,5.9)
```

```python
p1.show_info()
```

```python
class Employee:
    company = "ABC Technologies"
    def __init__(self,name,monthly_salary,job_designation):
        self.name=name
        self.monthly_salary=monthly_salary
        self.job_designation=job_designation
    def annual_salary(self):
        return self.__*12 #cprivate variable
    def show_info(self):
        return f"Name -> {self.name}, monthly_salary -> {self.monthly_salary}, job_designation -> {self.job_designation}, company -> {self.company}"
```

```python
e1=Employee("Jatin", 30000, "Developer")
e2=Employee("Sneha", 40000, "Manager")
```

```python
e1.show_info()
```
