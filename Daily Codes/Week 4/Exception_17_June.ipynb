```python
try:
    f=open("a1.txt")
    print(var1)
except:
    print("File does not exist")
```

```python
# we need to be specific about what kind of error we want to catch
```

```python
try:
    f=open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File does not exist")
```

```python
try:
    f=open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File does not exist")
except Exception:
    print('Something went wrong')
```

```python
try:
    f=open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File does not exist")
except Exception as e:
    print(e)
```

```python
# else --> 
```

```python
try:
    f=open('a1.txt')
except FileNotFoundError:
    print("File does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
```

```python
# finally
```

```python
try:
    f=open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("Run complete")
```

```python
try:
    f=open('a1.txt')
    if f.name=='a1.txt':
        raise Exception
except FileNotFoundError:
    print("File does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("Run complete")
```

```python
# take 2 numbers as user input
# write a program for division of these 2 numbers
# handle ZeroDivisionError and ValueError
```

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    result = num1 / num2
    print("Division =", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Values not entered properly.")
```
